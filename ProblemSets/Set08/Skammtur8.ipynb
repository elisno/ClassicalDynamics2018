# Lausn

TÃ¡knum innfallshorniÃ° meÃ° $\alpha$ og ÃºtfallshorniÃ° meÃ° $\theta$. TÃ¡knum upphafsferÃ°ina meÃ° $v_0$ og lokaferÃ°ina meÃ° $v$.

EingÃ¶ngu hraÃ°aÃ¾Ã¡tturinn Ã­ $x$-stefnu breytist Ã¾ar sem Ã¾verill snertiflatanna liggur eftir $x-Ã¡s$.
ÃvÃ­ er $v_y = v_{y0} = v_0 \sin\alpha$ og $v_x = \epsilon v_{x0} = \epsilon v_0 \cos\alpha$.

LokaferÃ°in er Ã¾vÃ­

$$

$$

ÃtfallshorniÃ° er

$$

```python
import numpy as np

v0 = 5.0  # m/s
epsilon = 0.8  
alpha = 30 * np.pi / 180.0 # rad

v = v0 * np.sqrt(epsilon**2 * np.cos(alpha)**2 + np.sin(alpha)**2)
theta = np.arctan(np.tan(alpha)/epsilon)

theta = theta * 180.0 / np.pi # deg

print("v =",round(v,2),"m/s")
print("theta =",round(theta,1),"Â°")

```
