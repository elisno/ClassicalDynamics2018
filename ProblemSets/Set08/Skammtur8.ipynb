A chain such as the one in Example 9.2 (with the same parameters) of length $b$ and mass $\rho b$ is suspended from one end at a point that is a height $b$ above a table so that the free end barely touches the tabletop. At time $t= 0$, the fixed end of the chain is released. Find the force that the tabletop exerts on the chain after the original fixed end has fallen a distance $x$.

# Lausn

Þegar að öll keðjan hefur fallið niður á borðið, þarf krafturinn frá borðinu á keðjuna að jafngilda þyngdarkraftinum á keðjuna ($F = Mg = \rho b g$).
Á meðan keðjan er í falli þarf krafturinn $F$ að stytta út þyngdarkraft þess hluta keðjunnar sem er nú þegar lentur, og einnig að "stela" skriðþunga fallandi hlutans.

$$
\frac{dp}{dt} = \rho b g - F = -\rho \dot{x}^2 + \rho (b-x)\ddot{x}
$$

Fyrra samasem merkið samsvarar skilgreiningu heildarkrafts á keðjuna og seinna samasem merkið samsvarar tímaafleiðunni af skriðþunga fallandi hluta keðjunnar $p = \rho (b-x)\dot{x}$.

Hér gerum við ráð fyrir að $\ddot{x} = g$ þar sem efsti endi keðjunnar verður eingöngu fyrir þyngdarkrafti.
Sá endi fær hreyfiorkuna $\frac{1}{2}dm \dot{x}^2$ með því að tapa stöðuorkunni $dm g x$.
Því er $\dot{x}^2 = 2gx$.

Við getum einangrað kraftinn $F$ úr jöfnunni fyrir heildarkraftinn og fáum

$$
F = 3 \rho g  x
$$

# Dæmi 9-42

A steel ball of velocity 5 m/s strikes a smooth, heavy steel plate at an angle of 30° from the normal. If the coefficient of restitution is 0.8, at what angle and velocity does the steel ball bounce off the plate?

# Lausn

Táknum innfallshornið með $\alpha$ og útfallshornið með $\theta$. Táknum upphafsferðina með $v_0$ og lokaferðina með $v$.

Eingöngu hraðaþátturinn í $x$-stefnu breytist þar sem þverill snertiflatanna liggur eftir $x-ás$.
Því er $v_y = v_{y0} = v_0 \sin\alpha$ og $v_x = \epsilon v_{x0} = \epsilon v_0 \cos\alpha$.

Lokaferðin er því

$$
v = \sqrt{v_x^2 + v_y^2} = v_0 \sqrt{\epsilon^2 \cos^2 \alpha + \sin^2 \alpha}
$$

Útfallshornið er

$$
\theta = \tan^{-1} \left( \frac{v_y}{v_x} \right) = \tan^{-1} \left( \frac{\tan\alpha}{\epsilon} \right)
$$

```python
import numpy as np

v0 = 5.0  # m/s
epsilon = 0.8  
alpha = 30 * np.pi / 180.0 # rad

v = v0 * np.sqrt(epsilon**2 * np.cos(alpha)**2 + np.sin(alpha)**2)
theta = np.arctan(np.tan(alpha)/epsilon)

theta = theta * 180.0 / np.pi # deg

print("v =",round(v,2),"m/s")
print("theta =",round(theta,1),"°")

```

# Lausn

Hér er fallandi massinn breytilegur $\dot{p} = \dot{m}\dot{x} + m \ddot{x} = \rho \dot{x}^2 + \rho x \ddot{x} 
$

Einnig verkar nettóþyngdarkraftur á fallandi hluta reipisins: $\dot{p} = mg = \rho g x$.

Þessar jöfnur til saman gefa ólínulegu afleiðujöfnuna $\dot{x}^2 + x \ddot{x} = gx$.

Gerum ráð fyrir að lausn $v$ sé á veldisforminu $v  = k x^\gamma$.

Þá er afleiðujafnan á forminu:

$$
k^2 x^{2\gamma} + k^2 \gamma x^{2 \gamma} = g x \qquad \Rightarrow \qquad k^2 (1 + \gamma) x^{2\gamma} = gx 
$$

Veldisvísirinn gefur okkur að $\gamma = \frac{1}{2}$ og þá fæst að $k = \sqrt{\frac{2g}{3}}$ þar sem $k^2 (1+ \gamma) = g$.

Við höfum fundið lausn á afleiðujöfnunni:

$$
v = \sqrt{\frac{2gx}{3}} \qquad \Rightarrow \qquad a = v \color{blue}{\frac{dv}{dx}} = \sqrt{\frac{2gx}{3}} \cdot \color{blue}{\frac{1}{2\sqrt{\frac{2gx}{3}}} \cdot \frac{2g}{3}} = \frac{g}{3}
$$

# Dæmi 9-03

Find the center of mass of a uniformly solid cone of base diameter $2a$ and height $h$ and a solid hemisphere of radius $a$ where the two bases are touching.

# Lausn

Massamiðjur hvors forms fyrir sig eru á $z$-ás vegna samhverfu.

Finnum $z_i = \frac{1}{m_i} \int z dm_i = \frac{\rho_i}{m_i} \int z dV_i$ fyrir sitthvort formið.
Þetta eru rúmheildi, svo við þurfum að velja heppilegt hnitakerfi.

Fyrir keiluna er best að nota sívalningshnit (þar er $dV = r dr d\phi dz$)


Keilan hefur massann $\rho \int dV$

$$
z_\text{keila} = \frac{\int_0^{2\pi} \int_0^a \int_0^{h - \frac{hr}{a}} r z dz dr d\phi}{\int_0^{2\pi} \int_0^a \int_0^{h - \frac{hr}{a}} r dz dr d\phi} = \frac{\frac{h^2}{2} \int_0^a r \left(1- \frac{r}{a}\right)^2 dr}{h \int_0^a r \left(1 - \frac{r}{a}\right) dr} = \frac{\frac{h^2}{2} \left[ \frac{r^2}{2} - \frac{2r^3}{3a} + \frac{r^4}{4a^2}\right]_0^a}{h \left[ \frac{r^2}{2} - \frac{r^3}{3a}\right]_0^a} = \frac{\frac{h^2}{2} \frac{a^2}{12}}{h \frac{a^2}{6}} = \frac{h}{4}
$$

Fyrir hálfkúluna notum við kúluhnit (þar er $dV = r^2  \sin \theta dr d\theta d\phi$) þar sem $z = r \cos\theta$

$$
z_\text{hálfkúla} = \frac{\int_0^{2\pi} \int_0^a \int_\frac{\pi}{2}^\pi  r^3 \cos\theta\sin\theta d\theta dr d\phi}{\int_0^{2\pi}  \int_0^a \int_\frac{\pi}{2}^\pi r^2 \sin\theta  d\theta dr d\phi} = \frac{2\pi \int_\frac{\pi}{2}^\pi \cos\theta \sin\theta d\theta \int_0^a r^3 dr }{\frac{4\pi}{6} a^3} = \frac{-\frac{1}{2} \frac{a^4}{4}}{\frac{1}{3}a^3} = - \frac{3}{8} a
$$


$$
\boxed{z_\text{keila} = \frac{h}{4}   \ \ \quad , \ \ \quad z_\text{hálfkúla} = -\frac{3}{8}a }
$$

Síðan lítum við á sitthvort formið sem punktmassa. Massamiðjakerfisins er því:

$$
z_\text{CM} = \frac{V_\text{keila} z_\text{keila} + V_\text{hálfkúla} z_\text{hálfkúla}}{V_\text{keila} + V_\text{hálfkúla}}
$$

þar sem við höfum gert ráð fyrir að formin hafa sama þéttleika.

$$
z_\text{CM} = \frac{\frac{2\pi h a^2}{6} \cdot \frac{h}{4} - \frac{4\pi a^3}{6} \cdot \frac{3a}{8}}{\frac{2\pi h a^2}{6} + \frac{4\pi a^3}{6}} = \frac{h^2 - 3a^2}{4h + 8a}
$$
