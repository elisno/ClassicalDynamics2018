# Dæmi 7-04

A particle moves in a plane under the influence of a force $f = -Ar^{\alpha-1}$ directed toward the origin; $A$ and $a$ (> 0) are constants. Choose appropriate generalized coordinates, and let the potential energy be zero at the origin. Find the Lagrangian equations of motion. Is the angular momentum about the origin conserved? Is the total energy conserved?

# Lausn

Fyrst að krafturinn er gefinn í pólarhnitum, væri skynsamlegast að velja alhnitin þannig.

$$
U = - \int f dr = A \int r^{\alpha-1}dr = \frac{A}{\alpha}r^\alpha
$$

Notum pólhnit:
$$
T = \frac{1}{2}m (\dot{r}^2 + (r \dot{\theta})^2)
$$

Skrifum Lagrange-fallið:

$$
L = T-U = \frac{1}{2}m (\dot{r}^2 + (r \dot{\theta})^2) - \frac{A}{\alpha}r^\alpha
$$

Skrifum Euler-Lagrange hreyfijöfnurnar fyrir $r$ og $\theta$.

* $r$

$$
mr\dot{\theta}^2 - Ar^{\alpha-1} \frac{d}{dt} \left( m\dot{r} \right) = 0 \qquad \Rightarrow \qquad m \ddot{r} - m r \dot{\theta}^2 + A r^{\alpha - 1} = 0
$$


* $\theta$
$$
-\frac{d}{dt} \left( mr^2\dot{\theta}\right) = 0 \qquad \Rightarrow \qquad mr^2\dot{\theta} \equiv l = \text{fasti}
$$

Hverfiþunginn varðveitist ($L = \vert \vec{r} \times \vec{p} \vert = r\hat{r} \times m(\dot{r}\hat{r} + r\dot{\theta}\hat{\theta}) = mr^2\dot{\theta}$)


Endurritum hreyfijöfnuna fyrir $r$:

$$
m \ddot{r} - \frac{l^2}{mr^3} + A r^{\alpha - 1} = 0
$$

Skoðum heildarorkuna $E = T+U = \frac{1}{2}m (\dot{r}^2 + (r \dot{\theta})^2) + \frac{A}{\alpha}r^\alpha $

$$
\frac{dE}{dt} = m\ddot{r}\dot{r} - \color{red}{\frac{l^2}{mr^3}}\dot{r} + A r^{\alpha -1}\dot{r} = 0 
$$

þar sem $\color{red}{\frac{1}{2}m (r\dot{\theta})^2 = \frac{l^2}{2mr^2}}$. Heildartimaafleiðan er núll því við getum deilt jöfnunni með $\dot{r}$ og fengið endurrituðu hreyfijöfnuna fyrir $r$.


Heildarorkan varðveitist því líka.

# Dæmi 7-05

Consider a vertical plane in a constant gravitational field. Let the origin of a coordinate system be located at some point in this plane. A particle of mass m moves in the vertical plane under the influence of gravity and under the influence of an additional force $f = -Ar^{\alpha-1}$ directed toward the origin ($r$ is the distance from the origin; $A$ and $a$ [ $\neq $ 0 or 1] are constants). Choose appropriate generalized coordinates, and find the Lagrangian equations of motion. Is the angular momentum about the origin conserved? Explain.

# Lausn


Hreyfiorkan hefur sama form, en stöðuorkan er $U = \frac{A}{\alpha}r^{\alpha} + mg r\sin\theta$.

$$
L = \frac{1}{2}m (\dot{r}^2 + (r \dot{\theta})^2) - \frac{A}{\alpha}r^\alpha - mg r\sin\theta
$$

* $r$

$$
m \ddot{r} - mr\dot{\theta}^2 + A r^{\alpha - 1} + mg\sin\theta = 0
$$


* $\theta$

$$
mgr \cos\theta - \frac{d}{dt}(mr^2\dot{\theta}) = 0 \qquad \Rightarrow \qquad mr^2\ddot{\theta} +  2mr\dot{r}\dot{\theta} - mgr \cos\theta  = 0
$$

# Dæmi 7-08

Consider a region of space divided by a plane. The potential energy of a particle in region 1 is $U_1$ and in region 2 it is $U_2$. If a particle of mass $m$ and with speed $v_1$ in region 1 passes from region 1 to region 2 such that its path in region 1 makes an angle $\theta_1$ with the normal to the plane of separation and an angle $\theta_2$ with the normal when in region 2, show that

\begin{equation}
\frac{\sin\theta_1}{\sin\theta_2} = \left( 1 + \frac{U_1 - U_2}{T_1} \right)^{1/2}
\end{equation}

where $T_1 = \frac{1}{2}mv_1^2$. What is the optical analog of this problem?



# Lausn

Veljum kartesísk hnit sem alhnit. Látum skilin á mættunum liggja á $y$-ás (þá má skrifa mættið $U$ sem fall af $x$ eingöngu).


$$
L = \frac{1}{2}m (\dot{x}^2 + \dot{y}^2) - U(x) 
$$

þar sem

$$
U(x) =
\begin{cases}
U_1 &, \text{ ef } x \lt 0 \\
U_2 &, \text{ ef } x \gt 0
\end{cases}
$$

Euler-Lagrange:

\begin{align}
m \ddot{x} + \frac{dU}{dx} &= 0\\
m \ddot{y} &= 0 \qquad \Rightarrow \qquad p_y = m\dot{y} = \text{ fasti}
\end{align}

Munum að $\ddot{x} = \frac{d v_x}{dx} v_x$. Þá fæst

$$
 \int_{v_{x_1}}^{v_{x_2}} v_x dv_x = - \frac{1}{m} \int_{U_1}^{U_2} dU
$$


$$
\Rightarrow \quad \frac{1}{2}m (v_{x_2}^2 - v_{x_1}^2) = - \frac{U_2 - U_1}{m}
$$

$$
\Rightarrow \quad \frac{1}{2}m v_{x_1}^2 + U_1 = \frac{1}{2}m v_{x_2}^2 + U_2
$$

Munum að $v_y$ er fasti: 

$$
\Rightarrow \quad \frac{1}{2}m v_1^2 + U_1 = \frac{1}{2}m v_2^2 + U_2 \qquad \text{ og } \qquad  v_1 \sin\theta_1 = v_2 \sin\theta_2
$$

Leysum saman:

$$
\frac{\sin\theta_1}{\sin\theta_2} = \frac{v_2}{v_1} = \frac{1}{v_1} \sqrt{v_1^2 + \frac{U_1 - U_2}{\frac{1}{2}m}} = \sqrt{1 + \frac{U_1 - U_2}{\frac{1}{2}mv_1^2}} = \sqrt{ 1 + \frac{U_1 - U_2}{T_1}}
$$


Þetta samsvarar Snell-lögmálinu um ljósbrot.


# Dæmi 7-16

The point of support of a simple pendulum of mass $m$ and length $b$ is driven horizontally by $ = a \sin \omega t$. Find the pendulum's equation of motion.

# Lausn


Lagrange-fallið hefur einfalt form í kartískum hnitum, en hreyfijöfnurnar einfaldast ef við skiptum yfir í pólhnit.

$$
L = \frac{1}{2}m(\dot{x}^2 + \dot{y}^2) + mgy  = \frac{1}{2}m((a\omega)^2 \cos^2(\omega t) + 2ab\omega\dot{\theta} \cos(\omega t) \cos\theta + (b\dot{\theta})^2) + mgb \cos\theta
$$

þar sem $x = a \sin(\omega t) + b\sin\theta$ og $y = - b \cos\theta$, o.s.frv.

Við byrjuðum með tvö hnit en endum í raun bara með eitt alhnit eftir ummyndunina í pólhnit. Við þurfum því bara að finna hreyfijöfnu fyrir $\theta$.

Euler-Lagrange gefur okkur að lokum:

$$
\ddot{\theta} + \frac{g}{b}\sin\theta - \frac{a \omega^2}{b}\sin(\omega t)\cos\theta = 0
$$

# Dæmi 7-01

A disk rolls without slipping across a horizontal plane. The plane of the disk remains vertical, but it is free to rotate about a vertical axis. What generalized coordinates may be used to describe the motion? Write a differential equation describing the rolling constraint. Is this equation integrable? Justify your answer by a physical argument. Is the constraint holonomic?

# Lausn

Hér þarf 4 alhnit til að lýsa hreyfingunni. T.d. $x$ og $y$ hnit fyrir stakan punkt á disknum, $\theta$ fyrir veltu disksins frá upphafs hornastöðu og $\phi$ sem lýsir spuna skífunnar um lóðréttan $z$ ás.

Það verkar núningur á snertiflöt disksins við planið sem leyfir honum bara að velta í eina stefnu.

$$
R d\theta = dx \cos\phi + dy \sin\phi
$$

Spunanum má lýsa með:

$$
\frac{dy}{dx} = \tan\phi
$$
Hvorug þessarra jafna er heildanleg, svo við höfum engar algebraískar jöfnur sem tengja öll hnitin saman.

Skorðurnar eru því *óheilnefndar*.

Það gæti hjálpað ef upphafsspunin væri þekktur eða að núningurinn á diskinn væri gefinn.

# Dæmi 7-06

A hoop of mass $m$ and radius $R$ rolls without slipping down an inclined plane of mass $M$, which makes an angle $\alpha$ with the horizontal. Find the Lagrange equations and the integrals of the motion if the plane can slide without friction along a horizontal surface.

# Lausn

Veljum kartesísk hnit fyrst fyrir massamiðju gjarðarinna út frá færslu skáplansins.

Látum $x = X + s\cos\alpha + R \sin \alpha$ og $y = R \cos\alpha + (L-S) \sin\alpha$,
þar sem $X$ er lárétt færsla skáplansins, $S$ er færsla gjarðarinnar eftir skáplaninu og $L$ er endanleg lengd skáplansins.

Við táknum veltu skífunnar með $\phi$, og veljum það þ.a. $S = R\phi$.

Orkur:

$$
L = T_\text{gjörð} + T_\text{skáplan} - U_\text{gjörð}= \frac{1}{2}m(2\dot{S}^2 + \dot{X}^2 + 2\dot{X}\dot{s}\cos\alpha) + \frac{1}{2}M \dot{X}^2 -  mg(R \cos\alpha + (L-S)\sin\alpha)
$$


Euler-Lagrange:

* $S$
$$
2m\ddot{S} + m \ddot{X}\cos\alpha - mg\sin\alpha = 0
$$

* $X$

$$
(m+M)\ddot{X} + m \ddot{S}\cos\alpha = 0
$$

Út frá seinni jöfnuna fæst að heildarskriðþungi kerfisins í lárétta stefnu varðveitist (því Lagrangefallið er óháð $X$).


Við getum einangrað afleiðurnar í báðum jöfnunum:

$$
 \ddot{S} = \frac{g \sin\alpha}{\left( 2 - \frac{m \cos^2\alpha}{m+M}\right)}
$$

$$
\ddot{X} = - \frac{mg \sin\alpha \cos\alpha}{2(m+M) - m \cos^2\alpha}
$$
