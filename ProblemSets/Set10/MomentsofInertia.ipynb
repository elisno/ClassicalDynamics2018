# Hverfitregða hálfkúlu

Reiknum hverfitregðu hálfkúlu um samhverfuás ($z$-ás). Gerum ráð fyrir að hálfkúlan sé einsleit með massann $m$ og radíus $b$.

$$
I_z = \int_0^{2\pi} \int_0^{\pi/2} \int_0^{b} \rho (x^2 + y^2) r^2 \sin\theta dr d\theta d\phi
$$

Í kúluhnitum er $x = r \sin \theta \cos \phi$ og $y = r \sin \theta \sin \phi$.
Svo er eðlismassi kúlunnar fasti ($\rho = \frac{m}{V} = \frac{m}{\frac{2}{3}\pi b^3}$).

$$
I_z = \frac{m}{\frac{2}{3}\pi b^3} \int_0^{2\pi} \int_0^{\pi/2} \int_0^{b} (r^2 \sin^2 \theta ) r^2 \sin\theta dr d\theta d\phi = \frac{2\pi m}{\frac{2}{3}\pi b^3} \cdot \frac{2}{3} \cdot \frac{b^5}{5} = \frac{2}{5}mb^2
$$

þar sem við notuðum að $\int_0^{\pi/2}\sin^3\theta d\theta = \int_0^{\pi/2} \frac{1}{4}(3\sin\theta - \sin3\theta)d\theta = \frac{2}{3}$ og $\int_0^b r^4 dr = \frac{b^5}{5}$


Vegna samhverfu vitum við að $I_x = I_y$ (eftir ásum sem fara í gegnum núllpunkt hálfkúlunnar).
Reiknum $I_x$ eins og við gerðum fyrir $I_z$:

$$
I_x = \frac{m}{\frac{2}{3}\pi b^3} \int_0^{2\pi} \int_0^{\pi/2} \int_0^{b} (r^2 \sin^2 \theta \sin^2\phi + r^2 \cos^2 \theta) r^2 \sin\theta dr d\theta d\phi
$$

Við klárum heildin yfir $r$ og $\theta$:

$$
I_x = \frac{mb^5}{5 \cdot \frac{2}{3}\pi b^3} \int_0^{2\pi} \int_0^{\pi/2} (\sin^3 \theta \sin^2\phi + \cos^2 \theta \sin\theta) d\theta d\phi
$$

$$
I_x = \frac{mb^5}{5 \cdot \frac{2}{3}\pi b^3} \int_0^{2\pi} \left(\frac{2}{3} \color{red}{\sin^2\phi} + \color{blue}{\int_0^{\pi/2}  \cos^2 \theta \sin\theta d\theta}  \right) d\phi
$$

$$
I_x = \frac{mb^5}{5 \cdot \frac{2}{3}\pi b^3} \left(\frac{2}{3} \color{red}{\pi} + \color{blue}{\frac{1}{3}} 2\pi  \right) = \frac{2}{5}mb^2
$$

og því $I_x = I_y = I_z = \frac{2}{5}mb^2$ fyrir hálfkúlu.


Það er fljótséð að utanhornalínustökin eru núll, þ.e.

$$
I_{ij} = - \frac{m}{\frac{2}{3}\pi b^3} \int_0^{2\pi} \int_0^{\pi/2} \int_0^{b} (x_i x_j)\sin\theta r^2 dr d\theta d\phi = 0
$$

Það er vegna þess að $\phi$-heildunin yfir $x_i x_j$ er aðgreinanleg og mun alltaf gefa núll (annað hvort heildi yfir $\cos\phi$ eða andstæðu föllin $\sin\phi$ og $\sin\phi\cos\phi$ á bilinu $0$ og $2\pi$).


$$
\mathbf{I}_0 = \frac{2}{5}mb^2 \begin{pmatrix}
1 & 0  & 0 \\ 
0 & 1  & 0 \\ 
0 & 0  & 1 
\end{pmatrix}
$$

# Hverfitregða tenings um massamiðju

Byrjum á hornalínustökunum (og heildum út frá massamiðjunni í staðin fyrir að heilda frá einu horni og nota reglu Steiner).

Vegna samhverfu um massamiðju þarf bara að reikna eitt hornalínustak.

$$
I_i = \frac{m}{l^3} \int_{-\frac{l}{2}}^{\frac{l}{2}} \int_{-\frac{l}{2}}^{\frac{l}{2}} \int_{-\frac{l}{2}}^{\frac{l}{2}} (x_j^2 + x_k^2) dx_i dx_j dx_k = \frac{m}{l^3} \left[\frac{x_i^3}{3}x_j + \frac{x_j^3}{3}x_i \right]_{-\frac{l}{2},-\frac{l}{2}}^{\frac{l}{2},\frac{l}{2}} \cdot \left[ x_k \right]_{-\frac{l}{2}}^\frac{l}{2} = \frac{m}{l^3} \left( \frac{l^3}{12}\cdot l + \frac{l^3}{12}\cdot l \right) \cdot l = \frac{1}{6}ml^2
$$

Reiknum utanhornalínustök:

$$
I_{ij} = -\int_{-\frac{l}{2}}^{\frac{l}{2}} \int_{-\frac{l}{2}}^{\frac{l}{2}} \int_{-\frac{l}{2}}^{\frac{l}{2}} (x_i x_j) dx_i dx_j dx_k = \left[\frac{x_i^2}{2} \frac{x_j^2}{2} x_k \right]_{-\frac{l}{2},-\frac{l}{2},-\frac{l}{2}}^{\frac{l}{2},\frac{l}{2},\frac{l}{2}} = 0
$$

vegna þess að heildin yfir $x_i$ og $x_j$ hvor um sig eru yfir oddstæð föll. ($\int_{-a}^a u du \cdot \int...= 0$) 

$$
\mathbf{I}_\text{CM}  = \frac{1}{6}mb^2 \begin{pmatrix}
1 & 0  & 0 \\ 
0 & 1  & 0 \\ 
0 & 0  & 1 
\end{pmatrix}
$$

# Hverfitregða disks um massamiðju

Látum diskinn vera sívalning með radíus $R$ og hæð $h$.
Notum sívalningshnit og látum $z$-ás vera samhverfuásinn. Byrjum á hornalínustökum hverfitregðuþinsins:

$$
I_z = \frac{m}{\pi R^2 h}\int_{0}^{2\pi} \int_{-\frac{h}{2}}^{\frac{h}{2}} \int_0^R (x^2 + y^2) r dr dz d\phi = \frac{m}{\pi R^2 h} \int_{0}^{2\pi} \int_{-\frac{h}{2}}^{\frac{h}{2}} \int_0^R r^3 dr dz d\phi = \frac{m}{\pi R^2 h} \cdot 2\pi \cdot h \cdot \frac{R^4}{4} = \frac{1}{2}mR^2
$$

Síðan höfum við að $I_x = I_y$ vegna samhverfu.

$$
I_x = \frac{m}{\pi R^2 h}\int_{0}^{2\pi} \int_{-\frac{h}{2}}^{\frac{h}{2}} \int_0^R (y^2 + z^2) r dr dz d\phi = \frac{m}{\pi R^2 h} \int_0^{2\pi} int_{-\frac{h}{2}}^{\frac{h}{2}} \int_0^R (r^3\cos^2\phi + z^2 r) dr dz d\phi =  \frac{m}{\pi R^2 h} \cdot (\frac{R^4}{4}\pi + \frac{h^3}{12}) = \frac{1}{12} m(3R^2 + h^2)
$$

Ákvörðum stök hverfitregðuþinsins sem eru utan hornalínunnar:

$$
I_{ij} = - \frac{m}{\pi R^2 h}\int_{0}^{2\pi} \int_{-\frac{h}{2}}^{\frac{h}{2}} \int_0^R  x_i x_j  r dr dz d\phi = 0
$$

Hvert stak hefur $\phi$-heildið $\int_0^{2\pi} \sin\phi d\phi=0$, $\int_0^{2\pi} \cos\phi d\phi=0$ eða $\int_0^{2\pi} \sin\phi \cos\phi d\phi = 0$.


$$
\mathbf{I}_\text{CM}  = \begin{pmatrix}
\frac{1}{12} m(3R^2 + h^2) & 0  & 0 \\ 
0 & \frac{1}{12} m(3R^2 + h^2)  & 0 \\ 
0 & 0  & \frac{1}{2}mR^2  
\end{pmatrix}
$$


Athugum að fyrir stöng með hverfandi radíus $R$ fæst:
$$
\mathbf{I}_\text{stöng,CM}  = \begin{pmatrix}
\frac{1}{12} mh^2 & 0  & 0 \\ 
0 & \frac{1}{12} mh^2  & 0 \\ 
0 & 0  & 0  
\end{pmatrix}
$$

Fyrir disk með hverfandi þykkt $h$ fæst:
$$
\mathbf{I}_\text{diskur,CM}  = \begin{pmatrix}
\frac{1}{4} mR^2 & 0  & 0 \\ 
0 & \frac{1}{4} mR^2  & 0 \\ 
0 & 0  & \frac{1}{2}mR^2  
\end{pmatrix}
$$
